```python
import keras
```

```python
from keras.models import Model
from keras.layers import Input, Dense
from keras.layers import Conv2D
from keras.layers import BatchNormalization
from keras.layers import Activation
```

```python
input_data = Input(shape=(19, 19, 17))
```

```python
def conv_block(x):
    y = Conv2D(256, (3, 3), padding='same')(x)
    y = BatchNormalization()(y)
    y = Activation('relu')(y)
    return y
```

```python
def residual_block(x):
    y = Conv2D(256, (3, 3), padding='same')(x)
    y = BatchNormalization()(y)
    y = Activation('relu')(y)
    y = Conv2D(256, (3, 3), padding='same')(y)
    y = BatchNormalization()(y)
    y = keras.layers.add([x, y])
    y = Activation('relu')(y)
    return y
```

```python
def policy_head(x):
    y = Conv2D(2, (1, 1), padding='same')(x)
    y = BatchNormalization()(y)
    y = Activation('relu')(y)
    y = Dense(19**2+1, activation='sigmoid')(y)
    return y
```

```python
def value_head(x):
    y = Conv2D(1, (1, 1), padding='same')(x)
    y = BatchNormalization()(y)
    y = Activation('relu')(y)
    y = Dense(256)(y)
    y = Activation('relu')(y)
    y = Dense(1)(y)
    y = Activation('tanh')(y)
    return y
```

```python
x = conv_block(input_data)

for i in range(39):
    x = residual_block(x)
    
policy_out = policy_head(x)
value_out = value_head(x)
```

```python
model = Model(inputs=[input_data], outputs=[policy_out, value_out])
```
